```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file with correct parsing
data = pd.read_csv(r"C:\Users\Nabeel Jawwad\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_6560597\API_SP.POP.TOTL_DS2_en_csv_v2_6560597.csv", skiprows=4)

# Display basic information about the DataFrame
print(data.info())

# Display the first few rows of the DataFrame
print(data.head())
```

```python
# Extract the relevant columns for visualization
population_data = data[['Country Name', '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']]

# Set 'Country Name' as the index
population_data.set_index('Country Name', inplace=True)

# Transpose the DataFrame to have years as rows and countries as columns
population_data = population_data.T

# Plot the population trend for selected countries
plt.figure(figsize=(12, 8))
for country in ['India', 'United States', 'China', 'United Kingdom', 'Brazil']:  # Example of selected countries
    plt.plot(population_data.index, population_data[country], label=country)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Trend Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate population growth rates for selected countries
population_growth_rates = population_data.pct_change() * 100

# Plot population growth rates for selected countries
plt.figure(figsize=(12, 8))
for country in ['India', 'United States', 'China', 'United Kingdom', 'Brazil']:  # Example of selected countries
    plt.plot(population_growth_rates.index, population_growth_rates[country], label=country)

plt.xlabel('Year')
plt.ylabel('Population Growth Rate (%)')
plt.title('Population Growth Rate Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
